```python
with open("input0303", "w") as f:
    r = 10
    c = 10
    x = 0
    y = 0
    f.write("{} {}\n".format(r, c))
    f.write("{} {}\n".format(x, y))
    for i in range(r):
        if i == 0:
            for j in range(c):
                if j == c - 1:
                    f.write("{}\n".format(1))
                else:
                    f.write("{} ".format(1))
        else:
            for j in range(c):
                if j == c - 1:
                    f.write("{}\n".format(1))
                elif j == 0:
                    f.write("{} ".format(1))
                else:
                    f.write("{} ".format(0))
```
